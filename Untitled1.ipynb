## Problem Statement
This program takes marks of multiple students as input,
stores them in a NumPy array, and calculates the total,
average, highest and lowest marks. It also assigns a
grade based on the average marks of all students.


## Logic Explanation
1. First I ask how many students there are
2. Then I use a for loop to collect each student's marks
3. I store marks in a list, then convert to NumPy array
4. I use NumPy functions to calculate total, average, highest, lowest
5. I defined a function assign_grade() that uses if/elif/else
   to return the grade based on average marks

##Import numpy as np:

##num_students:

Asks the user how many students there are input always gives text, so int converts it to a whole number.

##marks_list=[]:

Creates an empty box where marks will be stored one by one.

##loop to collect marks:

Repeats exactly num_students times. Each time it asks for one student's marks and adds it to the list using append.

##arr = np.array(marks_list)

Converts the Python list into a NumPy array so we can use NumPy's math functions on it.

##Lines 8-15:

Calculations

##Lines 16-23: Grade function

A function you defined yourself. It takes the average and checks conditions top to bottom, returning the matching grade.

##Print Results:

Calls your function with the calculated average and prints the final grade.

```python
import numpy as np
num_students = int(input("Enter number of students: "))
marks_list = []

for i in range(num_students):
    mark = int(input("Enter marks of student: "))
    marks_list.append(mark)


arr = np.array(marks_list)
print(f"Marks entered: {arr}")
total_marks = np.sum(arr)
print(f"The total marks of the students is {total_marks}")
average_marks = np.mean(arr)
print(f"The average marks of the students is {average_marks}")
highest_marks = np.max(arr)
print(f"The highest marks among students is {highest_marks}")
lowest_marks = np.min(arr)
print(f"The lowest marks among students is {lowest_marks}")

def assign_grade(avg_marks):
  if avg_marks >= 85:
      return "A-Grade"
  elif avg_marks >= 70:
      return "B-Grade"
  elif avg_marks >= 50:
      return "C-Grade"
  else:
      return "Fail"

grade = assign_grade(average_marks)
print(f"The overall grade for these marks is {grade}")
print("\n----- RESULT -----")
```
